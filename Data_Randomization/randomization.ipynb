```python
# In the following codes we are going to create a new column to our existing csv file.
# The new column (Continous_Assesments) will be containing the randomized marks for the 63 students
```

```python
# Importing the necessary packages
import pandas as pd
import numpy as np
import os
```

```python
# Assigning the csv file to tha dataframe data
data = pd.read_csv("New_File.csv")
```

```python
# Displaying the dataframe

data.head(10)
```

```python
# From the dataframe, lets rename the second column into something meaningfull

data = data.rename(columns={'Unnamed: 2' : 'Students_Name'})
```

```python
# Displaying the dataframe

data.head(7)
```

```python
# Adding a new column that will be containing randomized data for Continous_Assesment, lets assume the score was between 35 and 60

data['Continous_Assesment'] = np.random.randint (35 , 57 ,size=(len(data)))
```

```python
data.head(27)
```

```python
# Since the new data and changes made are not saved to the new file, lets save the new chages into a whole new csv file then

data.to_csv("CAs.csv" , index=False)
```
